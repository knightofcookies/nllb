```python
with open("../datasets/samanantar_4950k_filtered.tsv", "r", encoding="utf-8") as fp:
    lines = fp.readlines()[1:]
```

```python
kha_ltc = {} # ltc : length to count
en_ltc = {}

for line in lines:
    en_line, kha_line = line.split("\t\t\t\t\t")
    en_line_length = len(en_line.split(" "))
    kha_line_length = len(kha_line.split(" "))
    if en_line_length not in en_ltc:
        en_ltc[en_line_length] = 1
    else:
        en_ltc[en_line_length] += 1
    if kha_line_length not in kha_ltc:
        kha_ltc[kha_line_length] = 1
    else:
        kha_ltc[kha_line_length] += 1
```

```python
print(en_ltc)
```

```python
print(kha_ltc)
```

```python
l_5_count = 0
g_30_count = 0
for key in en_ltc:
    if key < 5:
        l_5_count += en_ltc[key]
    elif key > 30:
        g_30_count += en_ltc[key]
print(l_5_count, g_30_count)
```

```python
l_5_count = 0
g_30_count = 0
for key in kha_ltc:
    if key < 5:
        l_5_count += kha_ltc[key]
    elif key > 30:
        g_30_count += kha_ltc[key]
print(l_5_count, g_30_count)
```
